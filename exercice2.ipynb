```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
#to ignore warnings
import warnings
warnings.filterwarnings('ignore')

#import, read file then print its content
#find number of (rows, columns)
print("Data Info\n")
print("HEAD FOR: TRAIN.CSV\n ")
fileI = 'train.csv'
data_frame = pd.read_csv(fileI)
print(data_frame)
print("INFO FOR: GENGER SUMBISSION.CSV\n ")
print(data_frame.info())


#dropping rows with missing values (survived, sex or passenger class)
print("\nDropping rows with missing values\n")
miss_val = data_frame.dropna(subset=['Survived','Sex', 'Pclass'])
print(miss_val)

#plotting now 
print()
print("Survival by Gender\n")
plt.figure(figsize=(5,10))
sns.boxplot(x= 'Sex', y= 'Survived', data= miss_val)
plt.xlabel('Sex')
plt.ylabel('Survived')
plt.title('Survival by Gender')
plt.show()

print("Surbival by Passenger Class")
plt.figure(figsize=(5,10))
sns.boxplot(x= 'Pclass', y= 'Survived', data= miss_val)
plt.xlabel('Pclass')
plt.ylabel('Survived')
plt.title('Survival by Gender')
plt.show()



'''
print("\nStatistic summarys\n")
print(data_frame.describe().T)

print("\nFOR TEST.CSV\n")

fileII ='test.csv'
df2 = pd.read_csv(fileII)
print(df2.head)
print("Dropping rows with missing values")
print(df2.dropna())

'''
```
